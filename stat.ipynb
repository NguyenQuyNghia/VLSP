```python
import pandas as pd
import json

# Đọc dữ liệu từ tệp CSV
df = pd.read_csv('train.csv')

# Chuyển đổi chuỗi JSON thành đối tượng Python (dictionary)
df['quintuple'] = df['quintuple'].apply(lambda x: json.loads(x) if pd.notna(x) else None)

# Sử dụng pd.json_normalize để tách các keys thành các cột
df_normalized = pd.json_normalize(df['quintuple'])

# Kết hợp DataFrame gốc và DataFrame mới đã tách các keys
df = pd.concat([df.drop(['quintuple'], axis=1), df_normalized], axis=1)

df.head()
```

```python
print(f"so luong cau la {df['id'].nunique()}" )
print(f"so luong cau khong so sanh la {(df['comparison_sentence'] == 0).sum()}" )
df_comparison = df[df['comparison_sentence'] == 1]
print(f"so luong cau so sanh la {df_comparison['id'].nunique()}")
```
